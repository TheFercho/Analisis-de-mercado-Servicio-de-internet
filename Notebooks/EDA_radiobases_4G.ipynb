#### Radiobases por provinicia cada 100mil habitantes

```python
import pandas as pd
df_radiobases=pd.read_csv('./8_radiobases_4g_por_prov_100mil_hab_OK.csv')
```

```python
df_radiobases.info()
```

```python
df_radiobases.head(25)
```

```python
from matplotlib import pyplot as plt
X=df_radiobases['Provincia']
y=df_radiobases['Radiobases por cada 100 mil habitantes']
```

```python
plt.figure(figsize=(10,6))
plt.bar(X,y,label='Antenas por Provincia',color='orange')
plt.xticks(rotation=90)
plt.ylabel('Radiobases cada 100mil habitantes')
plt.legend(loc='upper center')  # Agregar leyenda
plt.tight_layout()  # Ajustar el espaciado
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

```python
df_radiobases['Radiobases por cada 100 mil habitantes'].describe()
```

Nota : Se puede verificar que: excluyendo a Capital Federal (CABA), que es claramente superior al promedio, el resto de las provincias pas se encuentran cerca del promedio general.
